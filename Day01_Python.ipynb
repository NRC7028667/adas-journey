```python
# CONCEPT 4 — STRING METHODS

sensor = "camera front sensor"

print(sensor.upper())               # CAMERA FRONT SENSOR
print(sensor.lower())               # camera front sensor
print(sensor.title())               # Camera Front Sensor
print(sensor.replace("front","rear")) # camera rear sensor
print(sensor.split(" "))            # ['camera', 'front', 'sensor']
print(len(sensor))                  # 19
```

```python
text = "   hello world   "

print(text.strip())          # removes spaces from both ends
print(text.strip().upper())  # chain methods together
print("hello".startswith("he"))  # True
print("hello".endswith("lo"))    # True
print("camera".count("a"))       # 2
print("hello world".find("world")) # 6 (position where word starts)
```

```python
# CONCEPT 5 — F-STRINGS

name     = "LiDAR"
distance = 25.678
active   = True

# Put f before the opening quote
# Put variable inside curly braces {}
print(f"Sensor: {name}")

# Control decimal places with :.2f
print(f"Distance: {distance:.2f} metres")   # 2 decimal places
print(f"Distance: {distance:.1f} metres")   # 1 decimal place
print(f"Distance: {distance:.0f} metres")   # 0 decimal places

# Multiple variables in one line
print(f"{name} is {distance:.2f}m away. Active: {active}")
```

```python
frame  = 7
name   = "Car"
dist   = 14.3

# :04d means show as 4 digits with leading zeros
print(f"Frame: {frame:04d}")   # Frame: 0007

# :10s means pad string to 10 characters
print(f"Object: {name:10s} at {dist:.2f}m")  # Object: Car        at 14.30m

# Show a calculation inside the curly braces
speed = 10
print(f"TTC: {dist/speed:.2f} seconds")    # TTC: 1.43 seconds
```

```python
# TASK 5

sensor   = "Ultrasonic"
reading  = 7.8934
frame_id = 3

# Using f-strings print:
# 1. "Sensor: Ultrasonic"
# 2. "Reading: 7.89 metres"  (2 decimal places)
# 3. "Frame: 0003"           (4 digits with leading zeros)
# 4. "Ultrasonic reading is 7.9m on frame 0003"
#    (1 decimal place, frame 4 digits)

# Write below:
print(f"Sensor:{sensor}\nReading:{reading:.2f}\nFrame:{frame_id:04d}")
print(f"Sensor {sensor} Reading is {reading:.1f} on Frame {frame_id:04d}")
```

```python
# CONCEPT 6 — BOOLEANS AND COMPARISONS

distance = 25.0

# Comparison operators
print(distance > 20)     # True
print(distance < 20)     # False
print(distance >= 25)    # True
print(distance <= 24)    # False
print(distance == 25.0)  # True
print(distance != 30.0)  # True
```

```python
# and — both must be True
# or  — at least one must be True
# not — flips True to False and False to True

camera_ok = True
radar_ok  = False
lidar_ok  = True

print(camera_ok and lidar_ok)    # True  — both True
print(camera_ok and radar_ok)    # False — radar is False
print(camera_ok or  radar_ok)    # True  — camera is True
print(radar_ok  or  False)       # False — neither True
print(not camera_ok)             # False — flips True
print(not radar_ok)              # True  — flips False
```

```python
# TASK 6

speed       = 15.0
speed_limit = 30.0
camera_on   = True
radar_on    = False
lidar_on    = True

# Print True or False for each:
# 1. Is speed less than speed_limit?
# 2. Is speed greater than 20?
# 3. Is speed exactly equal to 15.0?
# 4. Are both camera and lidar on?
# 5. Is at least one sensor on?
# 6. Is radar NOT on?
# 7. Are all three sensors on?

# Write below:
print( speed < speed_limit)
print(speed > 20)
print(speed == 15)
print(camera_on and lidar_on)
print(camera_on or lidar_on or radar_on)
print(not radar_on)
print(camera_on and lidar_on and radar_on)
```

```python
# In ADAS — boolean checks decide system actions

ttc           = 1.4    # time to collision in seconds
sensor_ok     = True   # sensor is working
speed_ms      = 15.0   # current speed

# Check if AEB should fire
aeb_should_fire = ttc < 1.6 and sensor_ok and speed_ms > 5.0

print(f"AEB should fire: {aeb_should_fire}")   # True

# All three conditions must be True for AEB to fire
# ttc < 1.6 → True
# sensor_ok → True
# speed > 5 → True
# So AEB fires
```

```python
# CONCEPT 7 — PRINT IN DETAIL

# Basic print
print("Hello")

# Print multiple values — Python adds space between them
print("Distance:", 25.7, "metres")

# sep — change what goes between values
print("Camera", "Radar", "LiDAR", sep=" | ")
# Output: Camera | Radar | LiDAR

print("Camera", "Radar", "LiDAR", sep=", ")
# Output: Camera, Radar, LiDAR

# end — change what goes at the end (default is newline)
print("Loading", end="")
print("...", end="")
print(" Done")
# Output: Loading... Done  (all on one line)

# Print empty line
print()

# Print a separator line
print("=" * 40)
```
