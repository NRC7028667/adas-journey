```python
# CONCEPT 1 — FUNCTIONS BASICS

# def keyword creates a function
# show_message is the name
# () means no inputs needed
# colon starts the function body
# everything indented belongs to the function
def show_message():
    print("Hello from the function")
    print("This runs every time you call the function")

# The function is defined but has NOT run yet
# To run it you CALL it by writing its name with ()
show_message()
print()
show_message()
print()
show_message()
```

1. Python reads def show_message(): — stores the function in memory
2. Python sees show_message() — goes to that stored function
3. Runs the code inside
4. Comes back and continues
5. sees show_message() again — repeats

```python
# CONCEPT 2 — FUNCTIONS WITH PARAMETERS

# 'name' is a parameter — it is like a variable
# that only exists inside this function
def greet(name):
    print(f"Hello {name}")

# When you call the function you give it a value
# That value is called an argument
greet("Camera")
greet("Radar")
greet("LiDAR")
```

```python
# Functions can take more than one parameter
# Separate them with commas

def describe_sensor(name, max_range, active):
    print(f"Sensor : {name}")
    print(f"Range  : {max_range} metres")
    print(f"Active : {active}")
    print()

# Pass arguments in the same order as parameters
describe_sensor("Camera", 250, True)
describe_sensor("Radar",  200, True)
describe_sensor("LiDAR",  150, False)
```

```python
# You can give a parameter a default value
# If the caller does not provide it — the default is used

def describe_sensor(name, max_range, active=True):
    print(f"{name} | Range: {max_range}m | Active: {active}")

# With all 3 arguments
describe_sensor("Camera", 250, False)

# Without the third — uses default True
describe_sensor("Radar", 200)

# Without third — uses default True
describe_sensor("LiDAR", 150)
```

```python
# CONCEPT 3 — RETURN VALUES

# Without return — function does something but gives nothing back
def add_no_return(a, b):
    result = a + b
    print(result)   # prints but gives nothing back

# With return — function gives the result back
def add_with_return(a, b):
    result = a + b
    return result   # sends result back to the caller

# Without return — you cannot use the result
x = add_no_return(3, 4)    # prints 7
print(x)                   # prints None — nothing was returned

print()

# With return — you can store and use the result
y = add_with_return(3, 4)  # result comes back and stored in y
print(y)                   # prints 7
print(y * 2)               # prints 14 — you can use it!
```

```python
# Python functions can return more than one value
# Separate them with commas

def get_min_max(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    return minimum, maximum    # returns two values

# Unpack into two variables
low, high = get_min_max([45.2, 12.1, 33.8, 8.7, 27.4])

print(f"Minimum: {low}")
print(f"Maximum: {high}")
```

```python
# TASK 3
# Write a function called calculate_ttc
# Parameters: distance, speed
# It should:
#   Return distance divided by speed
#   But if speed is 0 — return -1 (cannot divide by zero)
#
# Then call it with these values and print results:
# calculate_ttc(30.0, 10.0)  → should return 3.0
# calculate_ttc(15.0, 5.0)   → should return 3.0
# calculate_ttc(25.0, 0)     → should return -1

# Write below:

def calculate_ttc(distance, speed):
    if speed == 0:
        return -1
    return distance / speed

print(calculate_ttc(30.0, 10.0))
print(calculate_ttc(15.0, 5.0))
print(calculate_ttc(25.0, 0.0))
```

```python
# In ADAS — functions do one specific job
# This function decides AEB action from TTC value

def get_aeb_action(ttc):
    if ttc < 0:
        return "INVALID"
    elif ttc < 0.8:
        return "FULL BRAKE"
    elif ttc < 1.6:
        return "PARTIAL BRAKE"
    elif ttc < 2.5:
        return "WARNING"
    elif ttc < 3.5:
        return "PRE ALERT"
    else:
        return "ALL CLEAR"


# Use both functions together
def process_detection(distance, speed):
    ttc    = calculate_ttc(distance, speed)
    action = get_aeb_action(ttc)
    return ttc, action


# Test it
ttc, action = process_detection(12.0, 10.0)
print(f"TTC: {ttc:.2f}s | Action: {action}")

ttc, action = process_detection(5.0, 10.0)
print(f"TTC: {ttc:.2f}s | Action: {action}")

ttc, action = process_detection(45.0, 10.0)
print(f"TTC: {ttc:.2f}s | Action: {action}")
```

```python
# CONCEPT 4 — DICTIONARIES

# Curly braces create a dictionary
# Each entry is key: value
# Key is always a string (usually)
# Value can be any type

sensor = {
    "name"     : "RADAR_FRONT",
    "range"    : 200,
    "active"   : True,
    "distance" : 34.5
}

# Access a value using its key in square brackets
print(sensor["name"])       # RADAR_FRONT
print(sensor["range"])      # 200
print(sensor["active"])     # True
print(sensor["distance"])   # 34.5
```

```python
sensor = {
    "name"  : "Camera",
    "range" : 250,
}

print(sensor)   # {'name': 'Camera', 'range': 250}

# Add a new key
sensor["active"] = True
print(sensor)   # now has 3 keys

# Change an existing value
sensor["range"] = 300
print(sensor)   # range changed to 300

# Delete a key
del sensor["active"]
print(sensor)   # active removed

# Check if key exists
print("name" in sensor)    # True
print("speed" in sensor)   # False
```

```python
# CONCEPT 5 — DICTIONARY METHODS

sensor = {
    "name"     : "LiDAR",
    "range"    : 150,
    "active"   : True,
    "distance" : 22.5
}

# Get all keys
print(sensor.keys())
# dict_keys(['name', 'range', 'active', 'distance'])

# Get all values
print(sensor.values())
# dict_values(['LiDAR', 150, True, 22.5])

# Get all key-value pairs as tuples
print(sensor.items())

# Get a value safely — no error if key missing
print(sensor.get("name"))       # LiDAR
print(sensor.get("speed"))      # None — key does not exist
print(sensor.get("speed", 0))   # 0 — default value if key missing

# Number of keys
print(len(sensor))   # 4
```

```python
# CONCEPT 6 — LOOPS WITH DICTIONARIES

sensor = {
    "name"     : "Radar",
    "range"    : 200,
    "active"   : True,
    "distance" : 34.5
}

# Loop through keys only
print("Keys:")
for key in sensor:
    print(key)

print()

# Loop through keys and values together
# .items() gives you both at the same time
print("Keys and Values:")
for key, value in sensor.items():
    print(f"  {key:10s} : {value}")
```

```python
# Very common pattern in ADAS
# A list where each item is a dictionary
# Each dictionary = one detected object

detections = [
    {"class": "Car",        "distance": 35.2, "confidence": 0.92},
    {"class": "Pedestrian", "distance": 8.7,  "confidence": 0.85},
    {"class": "Cyclist",    "distance": 14.3, "confidence": 0.78},
]

print("=== All Detections ===")

for obj in detections:
    name = obj["class"]
    dist = obj["distance"]
    conf = obj["confidence"]
    print(f"{name:12s} | {dist:.1f}m | Conf: {conf:.2f}")
```

```python
# TASK 6
sensors = [
    {"name": "Camera",     "range": 250, "active": True},
    {"name": "Radar",      "range": 200, "active": True},
    {"name": "LiDAR",      "range": 150, "active": False},
    {"name": "Ultrasonic", "range": 8,   "active": True},
]

# Write a for loop that prints:
# "OK     : Camera     | Range: 250m"   for active sensors
# "FAULT  : LiDAR      | Range: 150m"   for inactive sensors
#
# Use f-string with name padded to 12 characters

# Write below:

for obj in sensors:
    status = "OK" if obj["active"] else"FAULT"

    print(
        f"{status:<7}: {obj['name']:<12}  Range: {obj['range']}m"
    )
```

```python
# CONCEPT 7 — PUTTING IT ALL TOGETHER
# A small ADAS detection processor

# ── STEP 1: Define helper functions ─────────────────────────

def calculate_ttc(distance, speed):
    if speed <= 0:
        return -1
    return distance / speed


def get_risk_level(distance, ttc):
    if distance < 10 or ttc < 1.6:
        return "HIGH"
    elif distance < 20 or ttc < 2.5:
        return "MEDIUM"
    else:
        return "LOW"


def print_detection(obj):
    name  = obj["class"]
    dist  = obj["distance"]
    speed = obj.get("speed", 10.0)    # default 10 m/s if not given
    conf  = obj["confidence"]

    ttc   = calculate_ttc(dist, speed)
    risk  = get_risk_level(dist, ttc)

    print(f"  {name:12s} | "
          f"Dist: {dist:5.1f}m | "
          f"TTC: {ttc:.2f}s | "
          f"Risk: {risk:6s} | "
          f"Conf: {conf:.2f}")


# ── STEP 2: Create detection data ───────────────────────────

frame_detections = [
    {"class": "Car",        "distance": 35.2, "confidence": 0.92, "speed": 8.0},
    {"class": "Pedestrian", "distance": 8.7,  "confidence": 0.85, "speed": 3.0},
    {"class": "Cyclist",    "distance": 14.3, "confidence": 0.78, "speed": 5.0},
    {"class": "Car",        "distance": 52.1, "confidence": 0.91, "speed": 12.0},
    {"class": "Truck",      "distance": 5.2,  "confidence": 0.88, "speed": 15.0},
]


# ── STEP 3: Process all detections ──────────────────────────

print("=" * 65)
print("  ADAS FRAME DETECTION REPORT")
print("=" * 65)

high_risk_count = 0

for obj in frame_detections:
    print_detection(obj)

    # Count high risk objects
    dist  = obj["distance"]
    speed = obj.get("speed", 10.0)
    ttc   = calculate_ttc(dist, speed)
    risk  = get_risk_level(dist, ttc)

    if risk == "HIGH":
        high_risk_count += 1

print("=" * 65)
print(f"  Total detections : {len(frame_detections)}")
print(f"  High risk objects: {high_risk_count}")

if high_risk_count > 0:
    print("  *** AEB SYSTEM ACTIVE ***")
else:
    print("  System nominal")

print("=" * 65)
```

```python
def summarise_sensors(sensor_list):
    # Total sensors
    total = len(sensor_list)

    # Active and inactive counts
    active = 0
    inactive = 0

    for sensor in sensor_list:
        if sensor["active"]:
            active += 1
        else:
            inactive += 1

    # Find sensor with highest range
    max_sensor = sensor_list[0]

    for sensor in sensor_list:
        if sensor["range"] > max_sensor["range"]:
            max_sensor = sensor

    # Print report
    print("=== Sensor Summary ===")
    print(f"Total    : {total}")
    print(f"Active   : {active}")
    print(f"Inactive : {inactive}")
    print(f"Max range: {max_sensor['name']} ({max_sensor['range']}m)")


# Call function
summarise_sensors(sensors)
```
