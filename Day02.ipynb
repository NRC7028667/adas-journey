```python
# elif means "else if"
# Check another condition if the first one was False

x = 10

if x > 10:
    print("greater than 10")
elif x == 10:
    print("exactly 10")
else:
    print("less than 10\n\n")
```

Indentation means the 4 spaces before the print statement.

```python
# In ADAS — loop through all detections in one frame
# and check each one

detections = [
    {"class": "Car",        "distance": 35.2},
    {"class": "Pedestrian", "distance": 8.7},
    {"class": "Cyclist",    "distance": 14.3},
    {"class": "Bike",        "distance": 52.1},
]

print("=== Frame Detection Report ===")

for obj in detections:
    name = obj["class"]
    dist = obj["distance"]
    ttc  = dist / 10.0    # assume 10 m/s closing speed

    if dist < 15:
        status = "DANGER"
    else:
        status = "CLEAR"

    print(f"{status}  | {name:12s} | {dist:.1f}m | TTC:{ttc:.1f}s")
```

```python
# CONCEPT 3 — WHILE LOOPS

count = 0

while count < 5:
    print(f"Count is: {count}")
    count = count + 1

print("Loop finished")
```

```python
# break stops the loop immediately
# even if the condition is still True

count = 0

while count < 10:
    print(count)
    if count == 3:
        print("Stopped early")
        break
    count = count + 1
```

```python
# continue skips the rest of the current iteration
# and goes back to check the condition again

count = 0

while count < 6:
    count = count + 1
    if count == 3:
        continue        # skip printing when count is 3
    print(count)
```

```python
# TASK 3
# Write a while loop that:
# Starts with distance = 50
# Subtracts 7 each loop
# Prints the distance each time
# Stops when distance goes below 10

# Write below:

dist = 50
while dist > 10:
    print(f"Distance:{dist}")
    dist = dist - 7 
```

```python
# In ADAS — simulate a vehicle approaching
# Loop runs until brakes fire or collision

distance  = 50.0
speed     = 12.0
time_step = 0.1

print("=== AEB Simulation ===")

while distance > 0:
    distance = distance - (speed * time_step)
    ttc = distance / speed

    print(f"Distance: {distance:.1f}m | TTC: {ttc:.2f}s")

    if ttc < 1.6:
        print("EMERGENCY BRAKE FIRED")
        break
```

```python
# CONCEPT 4 — RANGE

# range(stop) — from 0 up to but NOT including stop
for i in range(5):
    print(i)
```

```python
# range(stop)
print("range(5):")
for i in range(5):
    print(i,)

print()   # new line

# range(start, stop)
print("range(2, 6):")
for i in range(2, 6):
    print(i,)
print()

# range(start, stop, step)
print("range(0, 10, 2):")
for i in range(0, 10, 2):
    print(i,)

print()

# Count backwards
print("range(10, 0, -1):")
for i in range(10, 0, -1):
    print(i,)
```

```python
# In ADAS — process frames one by one
# Frame number is important for logging

readings = [45.2, 42.1, 38.7, 35.2, 31.8, 28.1, 23.5]

print("=== Frame by Frame Processing ===")

for frame in range(len(readings)):
    dist   = readings[frame]
    ttc    = dist / 10.0

    print(f"Frame {frame:02d} | Distance: {dist:.1f}m | TTC: {ttc:.1f}s")
```

```python
# CONCEPT 5 — LISTS

# Creating a list — square brackets, comma separated
sensors = ["Camera", "Radar", "LiDAR", "Ultrasonic"]

# Access by index — counting starts at 0
print(sensors[0])    # Camera
print(sensors[1])    # Radar
print(sensors[-1])   # Ultrasonic — last item
print(sensors[-2])   # LiDAR — second from last
```

```python
numbers = [10, 20, 30, 40, 50, 60, 70]

# list[start:stop] — stop is NOT included
print(numbers[0:3])    # [10, 20, 30]
print(numbers[2:5])    # [30, 40, 50]
print(numbers[:3])     # [10, 20, 30] — from beginning
print(numbers[4:])     # [50, 60, 70] — to the end
print(numbers[::2])    # [10, 30, 50, 70] — every 2nd item
print(numbers[::-1])   # [70, 60, 50, 40, 30, 20, 10] — reversed
```

```python
# CONCEPT 6 — LIST METHODS

my_list = [30, 10, 50, 20, 40]

# Add item to end
my_list.append(60)
print(my_list)          # [30, 10, 50, 20, 40, 60]

# Remove specific item
my_list.remove(10)
print(my_list)          # [30, 50, 20, 40, 60]

# Sort in ascending order
my_list.sort()
print(my_list)          # [20, 30, 40, 50, 60]

# Reverse the list
my_list.reverse()
print(my_list)          # [60, 50, 40, 30, 20]

# Find index of an item
print(my_list.index(40))   # 2

# Count how many times item appears
nums = [1, 2, 2, 3, 2, 4]
print(nums.count(2))       # 3

# Length of list
print(len(my_list))        # 5
```

```python
# CONCEPT 7 — NESTED LISTS

# A list inside a list
# Think of it as rows and columns
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Access one row
print(matrix[0])        
print(matrix[1])        
print(matrix[2])

# Access one specific item
# matrix[row][column]
print(matrix[0][0])     # 1 — row 0, column 0
print(matrix[1][2])     # 6 — row 1, column 2
print(matrix[2][1])     # 8 — row 2, column 1
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Loop through rows
for row in matrix:
    print(row)

print()

# Loop through every single item
for row in matrix:
    for item in row:
        print(item, end=" ")
    print()    # new line after each row
```

```python
# TASK 7 — FINAL CHALLENGE
#
# You have sensor readings for 3 sensors across 4 frames
# Each inner list = one sensor's readings across 4 frames

sensor_data = [
    [45.2, 42.1, 38.7, 35.2],   # Camera readings
    [44.8, 41.9, 39.1, 34.8],   # Radar readings
    [45.0, 42.5, 38.9, 35.0],   # LiDAR readings
]

sensor_names = ["Camera", "Radar", "LiDAR"]

# Do these tasks:
# 1. Print all readings for Camera (first inner list)
# 2. Print the Radar reading at frame 2 (index 2)
# 3. Loop through sensor_data and print each sensor name
#    with all its readings on one line
# 4. Find and print the minimum reading across ALL sensors
#    (hint: you need a loop inside a loop)

# Write below:
print("Camera reading:", sensor_data[0])
print("Radar Frame 2:",sensor_data[1][2])

for i in range(len(sensor_data)):
    print(sensor_names[i] + ":", sensor_data[i])

minimum = sensor_data[0][0]
for sensor in sensor_data:
    for reading in sensor:
        if reading < minimum:
            minimum = reading
print("Minimum reading across all sensors:", minimum)
```
